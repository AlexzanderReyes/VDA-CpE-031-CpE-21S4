c. Read all product sales data and show it using a multiline plot, Label them.

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('Technological-Products-Sample-Data.csv')
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(12, 6))

# Get unique brands and devices
all_brands = df['Brand'].unique()
all_devices = df['Device'].unique()

for brand in all_brands:
    brand_data = df[df['Brand'] == brand]

    # Plot each device for the current brand
    for device in brand_data['Device'].unique():
        device_data = brand_data[brand_data['Device'] == device]
        plt.plot(device_data['Device'], device_data['Sales'], marker='o', label=f"{brand} - {device}")

plt.xlabel('Device')
plt.ylabel('Sales (USD)')  # Corrected Y-axis label
plt.title('Product Sales Data by Brand and Device')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.legend(title='Brand - Device', bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside for better visibility
plt.grid(True)
plt.tight_layout()
plt.show()
```

f. Create a pie chart that shows each brand Country of Origin, Label them.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the data
data = pd.read_csv('Technological-Products-Sample-Data.csv')

# Count the number of brands from each country of origin
country_counts = data.groupby('Country of Origin')['Brand'].count()

# Create pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(country_counts, labels=country_counts.index,
                                  autopct='%1.1f%%', startangle=140,
                                  pctdistance=0.85)  # Adjust pctdistance for better readability

# Add brand information to the legend
legend_labels = []
for country, brand_count in country_counts.items():
    brands_in_country = data[data['Country of Origin'] == country]['Brand'].unique()
    brand_str = ', '.join(brands_in_country)
    legend_labels.append(f"{country} ({brand_str})")

plt.legend(wedges, legend_labels, title="Country of Origin (Brands)",
           loc="center left", bbox_to_anchor=(1, 0.5))

plt.title('Brand Distribution by Country of Origin')
plt.tight_layout()
plt.show()
```
