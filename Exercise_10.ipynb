```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

slope, intercept = np.polyfit(x,y,1)
y_pred = slope * x + intercept

plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
plt.show()
```

```python
x = np.linspace(-2,2,100)
y = x ** 2
fig = plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.show()
```

```python
# Data points
x = np.array([0.50, 1.00, 1.30, 1.60, 2.00, 2.20, 2.50, 3.10, 3.90, 4.40])
y = np.array([1.90, 1.50, 1.20, 1.00, 0.80, 0.78, 0.65, 0.46, 0.30, 0.23])

# Linear regression (slope and intercept)
slope, intercept = np.polyfit(x, y, 1)
y_pred = slope * x + intercept

# Scatter plot of original data
plt.scatter(x, y, label="Data Points", color="blue")

# Regression line
plt.plot(x, y_pred, color="red", label=f"y = {slope:.2f}x + {intercept:.2f}")

# Title and labels
plt.title("Scatter Plot of Correlation of Sample and Data Sets")
plt.xlabel("X")
plt.ylabel("Y")

# Legend
plt.legend()

# Show plot
plt.show()
```
