```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.50, 1.00, 1.30,1.60 ,2.00,2.20,2.50,3.10,3.90,4.40])
y = np.array([1.90,1.50, 1.20, 1.00, 0.80,0.78,0.65,0.46,0.30,0.23])

slope, intercept = np.polyfit(x, y, 1)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')


plt.title("Scatter PLot of Correlation of Sample and Data Sets")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

The linear regression shows a negative relationship between \(x\) and \(y\). As \(x\) increases, \(y\) decreases by a fixed amount, as indicated by the slope. The intercept shows the expected value of \(y\) when \(x = 0\). This means \(y\) gets smaller as \(x\) grows, and the regression line helps predict this pattern.
