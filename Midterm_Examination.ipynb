```python
import pandas as pd

file_path = "/content/HEARTDISEASE.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Question 1: Is there a correlation between sleep hours and the likelihood of having had a heart attack?

BAR PLOT

```python
heart_attack_data = df[df['HadHeartAttack'] == 'Yes']

heart_attack_counts = heart_attack_data['GeneralHealth'].value_counts()
total_counts = df['GeneralHealth'].value_counts()
heart_attack_percentage = (heart_attack_counts / total_counts) * 100

heart_attack_percentage.plot(kind='bar', color='red')
plt.title("Heart Attack Occurrence by General Health Status")
plt.xlabel("General Health Status")
plt.ylabel("Percentage of Heart Attack Cases")
plt.xticks(rotation=45)

plt.show()
```

The bar chart shows the percentage of people in each general health category (Excellent, Very Good, Fair) who have experienced a heart attack. We can observe if lower health ratings, like "Fair" or "Poor," have higher heart attack percentages compared to higher ratings, like "Excellent" or "Very Good." This helps us see how self-reported health status might relate to heart attack risk.

Question 2: What is the distribution of heart attack occurrences by sex?

PIE PLOT

```python
heart_attack_data = df[df['HadHeartAttack'] == 'Yes']

sex_distribution = heart_attack_data['Sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_distribution,labels=sex_distribution.index, autopct='%1.1f%%')

plt.title("Heart Attack Occurrences by Sex")
plt.axis('equal')

plt.show()
```

The pie chart illustrates the distribution of heart attack occurrences by sex, showing the percentage of males and females who have experienced a heart attack. If one gender has a significantly larger slice of the pie, it may indicate a higher prevalence of heart attacks among that group. This insight can help inform targeted prevention strategies and health interventions based on gender-specific risk factors for heart disease.
